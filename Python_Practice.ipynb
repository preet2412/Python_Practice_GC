#CPython typically caches integers from -5 to 256:


```
Small integers (-5 to 256) are cached in CPython.
None, True, and False are singletons.
Strings may be interned, but don't rely on it.  (Most of the time TRUE)
Floats, lists, dicts, sets, and most tuples are not cached in the same predictable way.

So type casting (int(), str(), float()) is a runtime operation performed
while the bytecode is being executed, not during bytecode generation.
✅ Small integers (-5 to 256) are pre-built at interpreter startup, not when you first use them.
✅ Python reuses those objects throughout the program.
❌ It is not "first time seen → then cached" for these small integers.
They already exist before your code starts running.
```

````python
import sqlite3 as s

con=s.connect("youtube.db")
cur=con.cursor()
cur.execute("""
CREATE TABLE IF NOT EXISTS YM (
id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT,
time REAL
)""")
con.commit()
def Db_show():
  cur.execute("SELECT * FROM YM")
  for i in cur.fetchall():
    print(i)

def Db_insert():
  name =input("Enter Name of Vedio pls :")
  time=input("Enter Time of Vedio pls :")
  cur.execute("INSERT INTO YM(name,time) VALUES(?,?)",(name,time))
  con.commit()

def Db_update():
  Db_show()
  name =input("Enter Name of Vedio pls Which YOU Want to Change :")
  time=input("Enter Time of Vedio pls which you want to Change :")
  cur.execute("UPDATE YM SET time=? WHERE name=?",(time,name))
  con.commit()

def Db_delete():
  Db_show()
  name =input("Enter Name of Vedio pls Which YOU Want to Change :")
  cur.execute("DELETE FROM YM WHERE name=?",(name,))
  con.commit()
def main():
  while(True):
    print("0 : Enter For Exit ")
    print("1 : Enter For See all Data ")
    print("2 : Enter For Insert Data ")
    print("3 : Enter For Update Data ")
    print("4 : Enter For Delete Data ")
    choice=input("Enter Your Choise Pls!")
    match choice:
      case "0" :
        print("You Are Exit Now......")
        break
      case "1":
        Db_show()
      case "2":
        Db_insert()
      case "3":
        Db_update()
      case "4":
        Db_delete()
    con.close()
if __name__=="__main__":
  main()

````

JWT contains:
1. User info (payload)
2. Signature

Signature is often created using HMAC + secret key.

That's why the server doesn't need a session table like cookies. It can recalculate the HMAC and verify the token.
Payload = Who are you?
Signature = Proof server created it.
