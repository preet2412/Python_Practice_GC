# Project 1 :Youtube (add time and name ) in file CRUD OPeration using JSON

```python
import json
def getvalue():
  n=input("enter you vedio name :")
  t=input("enter your time :")
  return {n:t}
def show_all():
  try:
    with open("Youtube_Project.txt","r") as f:
      print(json.load(f))
  except Exception as e:
    print(e)
  finally:
    pass

def insert_function(*args,**kwargs):
  z={}
  try:
    with open("Youtube_Project.txt","r") as f:
      z=json.load(f)
  except Exception as e:
    print(e)
  finally:
    pass
  key,value=iter(kwargs.items()).__next__()
  z[key]=value
  try:
    with open("Youtube_Project.txt","w") as f:
      json.dump(z,f)
  except Exception as e:
    print(e)
  finally:
    pass


def update_function(*args,**kwargs):
  with open("Youtube_Project.txt","r") as f:
        z=json.load(f)
  key,value=(iter(kwargs.items())).__next__()
  if key not in z:
    print("Not in your dictionary")
    return
  z[key]=value
  try:
    with open("Youtube_Project.txt","w") as f:
      json.dump(z,f)
  except Exception as e:
    print(e)
  finally:
    pass


def delete_function(*args,**kwargs):
   with open("Youtube_Project.txt","r") as f:
        z=json.load(f)
   key,value=iter(kwargs.items()).__next__()
   if key not in z:
    print("Not in your dictionary")
    return
   del z[key]
   try:
    with open("Youtube_Project.txt","w") as f:
      json.dump(z,f)
   except Exception as e:
    print(e)
   finally:
    pass


def __main__():
  while True:
    print("Enter 0 for Exit :")
    print("Enter 1 for Show ALL List :")
    print("Enter 2 for Insert :")
    print("Enter 3 for Update :")
    print("Enter 4 for Delete :")
    char=input("Enter Your Choice :")

    match char:
      case "0":
        print("You are Now Exit App")
        break
      case "1":show_all()
      case "2":
        value=getvalue()
        print(type(value))
        insert_function(**value)
      case "3":
        value=getvalue()
        print(value)
        show_all()
        update_function(**value)

      case "4":
        value=getvalue()
        show_all()
        delete_function(**value)
      case _:
        print("enter valid input pls !")
__main__()

```

# Project 2 (same Project 1 but with NOW real DATABSE SQLITE )

```python
import sqlite3 as s

con=s.connect("youtube.db")
cur=con.cursor()
cur.execute("""
CREATE TABLE IF NOT EXISTS YM (
id INTEGER PRIMARY KEY AUTOINCREMENT,
name TEXT,
time REAL
)""")
con.commit()
def Db_show():
  cur.execute("SELECT * FROM YM")
  for i in cur.fetchall():
    print(i)

def Db_insert():
  name =input("Enter Name of Vedio pls :")
  time=input("Enter Time of Vedio pls :")
  cur.execute("INSERT INTO YM(name,time) VALUES(?,?)",(name,time))
  con.commit()

def Db_update():
  Db_show()
  name =input("Enter Name of Vedio pls Which YOU Want to Change :")
  time=input("Enter Time of Vedio pls which you want to Change :")
  cur.execute("UPDATE YM SET time=? WHERE name=?",(time,name))
  con.commit()

def Db_delete():
  Db_show()
  name =input("Enter Name of Vedio pls Which YOU Want to Change :")
  cur.execute("DELETE FROM YM WHERE name=?",(name,))
  con.commit()
def main():
  while(True):
    print("0 : Enter For Exit ")
    print("1 : Enter For See all Data ")
    print("2 : Enter For Insert Data ")
    print("3 : Enter For Update Data ")
    print("4 : Enter For Delete Data ")
    choice=input("Enter Your Choise Pls!")
    match choice:
      case "0" :
        print("You Are Exit Now......")
        break
      case "1":
        Db_show()
      case "2":
        Db_insert()
      case "3":
        Db_update()
      case "4":
        Db_delete()
    con.close()
if __name__=="__main__":
  main()

```
