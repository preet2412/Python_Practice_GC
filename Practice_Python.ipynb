```python
from functools import reduce
print([i%2 for i in range(10) if i<8])
print('x' if 3<2 else 'z')
a=list(map(lambda x:x**2,range(10)))
p=[i**2 for i in range(10) if i%2==0]
print(p)
p=list(filter(lambda x:True if x<30 else False,p))
p=list(filter(lambda x:x<30,p)) #Shortcut
print(p)
p1=reduce(lambda x,y:x+y,p)
print(p1)
v=lambda x:(x/2)**2
print(v(8))
```
